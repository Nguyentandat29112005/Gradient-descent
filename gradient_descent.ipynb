# GRADIENT DESCENT TRONG LINEAR REGRESSION

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### TẢI CÁC THƯ VIỆN 

```python
data_frame = pd.read_csv("data.csv")
print(data_frame)
x = data_frame.iloc[:,0]
y = data_frame.iloc[:,1]
```

### ĐỊNH DẠNG DỮ LIỆU 

```python
plt.scatter(x,y)
plt.show()
```

### TẠO ĐỒ THỊ VỚI CÁC ĐIỂM DỮ LIỆU

```python
m = 0
c = 0 

L = 0.0001

n = float(len(x))


for i in range(10000):
    Y_pred = m*x +c
    D_m = (-2/n) * sum(x * (y-Y_pred))
    D_c = (-2/n) * sum(y-Y_pred)
    m = m - L*D_m
    c = c - L*D_c

print("Intercept = ", m)
print('Slope = ', c)

```

##### SỬ DỤNG GRADIENT DESCENT ĐỂ TÌM INTERCEPT VÀ SLOPE  
##### LEARNING RATE = 0.0001
##### THUẬT TOÁN GRADIENT DESCENT KHI CHẠY 10000 LẦN SẼ ĐƯA RA GIÁ TRỊ INTERCEPT VÀ SLOPE
#####


```python
Y_pred = m*x +c

plt.scatter(x,y)
plt.plot(x,Y_pred,color='red')
plt.show()
```
